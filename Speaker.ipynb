## Importing the relevant libraries

```python
import json
import csv
import pandas as pd
```

## Loading the JSON file

```python
# Opening JSON file and loading the data 
# into the variable data 
with open('speechToText.json') as json_file: 
    data = json.load(json_file) 
```

```python
data
```

## Retrieving the required list from the given dataset

```python
success = data['results'][0]['alternatives'][0]['timestamps']
success
```

## Converting the time stamp list and speaker list into dataframe and merging both

```python
df = pd.DataFrame(success,columns=['words','from','to'])
df
```

```python
sucess_2 = data['speaker_labels']
speaker_df = pd.DataFrame(sucess_2)
```

```python
speaker_df
```

```python
merge = pd.merge(df,speaker_df,how = "inner", on = ['from','to'] ) # merging on both from and to timestamp
```

```python
merge.head(50)
```

## Creating the list of speaker 1 and speaker 2 spoken words seperately using for and if loops

```python
sp1 = []
sp2 = []
for i, words in enumerate(merge.words):
    if merge.speaker[i] == 0:
        if sp1==[]:
            sp1.append([i,"Speaker 0 :", words])
        elif i == sp1[-1][0] + 1:
            a = sp1[-1][2]+" " + words
            sp1[-1][2] =a
            sp1[-1][0] = i
        else:
            sp1.append([i,"Speaker 0 :", words])
    elif merge.speaker[i] == 1:
        if sp2==[]:
            sp2.append([i,"Speaker 1 :", words])
        elif i == sp2[-1][0] + 1:
            a = sp2[-1][2]+" " + words
            sp2[-1][2] =a
            sp2[-1][0] = i
        else:
            sp2.append([i,"Speaker 1 :", words])
    
            
```

```python
sp1
```

```python
sp2
```

## Creating the dataframe of their conversation w.r.t to the time line

```python
sp = sp1 + sp2
df_sp = pd.DataFrame(sp,columns=['index','speaker','spoke'])
pd.set_option('display.max_colwidth', None)
df_sorted = df_sp.sort_values(by = 'index')
df_sorted[['speaker','spoke']]
```
